**Default Arguments**

```python
bill=eval(input('enter bill amount'))
tip_percent=eval(input('enter tip percentage'))
tip_Amount=bill*tip_percent/100
total_bill=bill+tip_Amount
print(f'Total bill amount is {total_bill}')
```

```python
#without arguments
def bill_pay():
    bill=eval(input('enter bill amount'))
    tip_percent=eval(input('enter tip percentage'))
    tip_Amount=bill*tip_percent/100
    total_bill=bill+tip_Amount
    print(f'Total bill amount is {total_bill}')

bill_pay()
```

```python
#with arguments
def bill_pay(bill,tip_percent):
    tip_Amount=bill*tip_percent/100
    total_bill=bill+tip_Amount
    print('Total bill amount is :',total_bill)

bill_pay(60,9)
```

```python
#using default argument
def bill_pay(bill,tip_percent=10):
    print('bill is :',bill)
    print('tip percent is :',tip_percent)
    tip_Amount=bill*tip_percent/100
    total_bill=bill+tip_Amount
    print('Total bill amount is :',total_bill)

bill_pay(10000,25)
```

**Default arguments always should be last**

```python
def average(n1,n3,n2=60):
    avg=(n1+n2+n3)/3
    print(round(avg,2))

average(30,40)
```

**Local variables** Global Variables**
    - Local variable means: variables declared inside the function call and can only be used inside the function
    - Global variable means: the variables declared outside the function and can be used inside and outside the function.

    

```python
#global s
s=0
def add():
    k=s+10
    s=c
    print(s)

add()
```

```python
s=0
def add():
    global c
    c=s+10
    print(c)

add()
```

```python
n1=10
def addition():
    n1=100
    n2=200
    n3=n1+n2
    print(n1,n2,n3)

addition()
```

```python
def add(a,b):
    print(a+b)

def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
    
def div(a,b):
    print(a/b)

def cal():
    print('enter 2 values')
    a=eval(input('enter first value'))
    b=eval(input('enter sec value'))
    option=eval(input('enter math option'))
    if option==1:
        add(a,b)
    elif option==2:
        sub(a,b)
    elif option==3:
        mul(a,b)
    elif option==4:
        div(a,b)

cal()
```
